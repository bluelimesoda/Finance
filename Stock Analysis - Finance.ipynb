```python
samsung = fdr.DataReader('005930','2018','2019')
```

```python
samsung.describe() #최소값 0의 의미는 주식 시장은 열였지만 거래가 없었던 날입니다. 
```

```python
fig, ax = plt.subplots(1,1, figsize = (7,7))
graph = sns.distplot(samsung['Close'], color='skyblue', 
                     label='Skewness: {:.2f}'.format(samsung['Close'].skew()), ax= ax)
graph = graph.legend(loc = 'best')
```

```python
sk =  fdr.DataReader('000660','2018','2019')
```

```python
sk.describe()
```

```python
fig, ax = plt.subplots(1,1, figsize = (7,7))
graph = sns.distplot(sk['Close'], color='pink', 
                     label='Skewness: {:.2f}'.format(sk['Close'].skew()), ax= ax)
graph = graph.legend(loc = 'best')
```

```python
fig, ax = plt.subplots(1,1)
ax = samsung['Close'].plot(figsize = (13,8), color = 'skyblue') 
ax = sk['Close'].plot(figsize = (13,8), color = 'pink')
plt.show() 
```

```python
fig = plt.figure(figsize = (12, 5))
ax = fig.add_subplot(111)
plt.plot(samsung.index, samsung['Volume'], label = 'Samsung')
plt.plot(sk.index, sk['Volume'], label = 'SK')
plt.xticks(rotation=70)
plt.legend(loc='best')
plt.show()
```

```python
cumax_samsung = samsung['Close'].cummax() #누적된 값 중에 가장 큰 값을 기준으로 주식의 추세 확인 
```

```python
cumax_samsung.plot(figsize = (12, 5))
```

```python
ax1 = plt.subplot(1,1,1)
cumax_samsung.plot(ax = ax1) 
samsung['Close'].plot(figsize = (12, 5), ax = ax1) #최대값 이후 주식 가격이 많이 내려감 
```

```python
cumax_sk = sk['Close'].cummax()
cumax_sk.plot(figsize = (12, 5))
```

```python
ax1 = plt.subplot(1,1,1)
cumax_sk.plot(ax = ax1) 
sk['Close'].plot(figsize = (12, 5), ax = ax1) #최대값 이후 종가가 많이 빠진 것을 볼수 있습니다. 
```

```python
sk['20d'] = np.round(sk['Close'].rolling(window = 20).mean(),2)
```

```python
sk['60d'] = np.round(sk['Close'].rolling(window = 60).mean(),2)
```

```python
sk['120d'] = np.round(sk['Close'].rolling(window = 120).mean(),2)
```

```python
sk['20-120'] = sk['20d'] - sk['120d']
sk['20-60']= sk['20d'] - sk['60d']
sk[['Close','20d','60d','120d']].plot(figsize = (20,5))
```

```python
SD = 65
sk['trading section'] = np.where(sk['20-120']> SD, 1, 0) 
```

```python
sk['trading section'] = np.where(sk['20-120'] < -SD, -1, sk['trading section']) 
```

```python
sk['trading section'].plot(lw=1.5)
```

```python
fig = plt.figure(figsize = (12,8))
ax3 = fig.add_subplot(111)
(sk['120d']/samsung['Close'].max()).plot(ax = ax3)
(sk['60d']/samsung['Close'].max()).plot(ax = ax3)
(sk['20d']/samsung['Close'].max()).plot(ax = ax3)
sk['trading section'].plot(ax = ax3)
plt.legend() # 이동평균과 매매구간을 하나의 그래프로 생성 
```
